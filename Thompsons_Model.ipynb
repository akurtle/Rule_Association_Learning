```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv("Ads_CTR_Optimisation.csv")
```

```python
import random
N = 10000
d = 10

ads_selected = []

number_of_rewards_1 = [0 for _ in range(d)]
number_of_rewards_0 = [0 for _ in range(d)]

total_reward = 0

for n in range(N):
  ad = 0
  max_random = 0
  for i in range(d):
    ran_beta = random.betavariate(number_of_rewards_1[i]+1 ,number_of_rewards_0[i]+1)
    if ran_beta>max_random:
      max_random = ran_beta
      ad = i

  ads_selected.append(ad)

  if(df.values[n,ad] == 1):
    number_of_rewards_1[ad]+=1
  else:
    number_of_rewards_0[ad]+=1

  total_reward += df.values[n,ad]
```

```python
plt.hist(ads_selected)
plt.title("Histogram of Ads selected")
plt.xlabel("Ads")
plt.ylabel("Number of times each ad was selected")
plt.show()
```
